```python
import pandas as pd
data = pd.read_csv("netflix_data.csv")
```

```python
data.tail()
```

```python
import numpy as np
data["cast"].isnull().sum()
```

```python
def len_cast (val):
    if val=="None":
        return 0
    else:
        return len(str(val).split(", "))
```

```python
data["cast"].replace(np.nan,"None",inplace=True)
```

```python
data.tail()
```

```python
data["len_cast"]=data["cast"].apply(len_cast)
```

```python
x= np.linspace(0,10,100)

plt.plot(x,np.sin(x), color="#000" ,linestyle= "solid" ,label="sin(x)")
plt.plot(x,np.sin(x-1),color="g",linestyle= "dashed" ,label="sin(x-1)")
plt.plot(x,np.sin(x-2),color="b",linestyle= "dotted" ,label="sin(x-2)")
plt.plot(x,np.sin(x-3),color="r",linestyle= "dashdot" ,label ="sin(x-3)")
plt.plot(x,np.sin(x-4), ":g" ,label="sin(x-4)")

plt.title("Sin Graph")
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.legend()
plt.show()
```
