```python
x=np.linspace(0,10,30)
y=np.sin(x)
plt.scatter(x,y)
plt.show()
```

```python
x=np.random.randn(100)

y=np.random.randn(100)

colors = np.random.rand(100)

sizes = 1000 * np.random.rand(100)

plt.scatter(x,y, c=colors,s=sizes ,alpha=0.3)

plt.colorbar()

plt.show()
```

```python
data= np.random.randn(100)
plt.hist(data)
plt.show()
```

```python
plt.hist(data, bins=50, alpha=0.5, histtype="stepfilled" ,color="steelblue")
plt.show()
```

```python
x1 = np.random.normal(0,0.8,1000)
x2 = np.random.normal(-2,1,1000)
x3 = np.random.normal(3,2,1000)

dict_extra = dict(histtype ="stepfilled", alpha =0.3, bins=30)

plt.hist(x1 , **dict_extra)
plt.hist(x2 , **dict_extra)
plt.hist(x3 , **dict_extra)

plt.show()
```

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x), '-b' ,label="Sine Graph")
plt.plot(x,np.cos(x), '--r' ,label="Cosine Graph")

plt.legend(loc='lower right' ,ncol=2)
plt.show()
```

```python
x,y = np.random.multivariate_normal([0,0],[[1,1],[1,2]],100000).T
plt.hist2d(x,y , bins=20 , cmap="Blues")
plt.show()
```

```python
x= np.linspace(0,10,100)
plt.plot(x, np.sin(x),label="Sine Graph")
plt.plot(x, np.cos(x),label="Cosine Graph")
plt.legend(loc="lower left", ncol=2, fancybox=True, borderpad=1, shadow = True)
plt.show()
```

```python
x = np.linspace(0,10,100)
inp =input("Enter function Sin or Cos")
if inp=="Sin":
    plt.plot(x, np.sin(x) ,label="Sin Graph")
else:
    plt.plot(x, np.cos(x), label="Cosine Graph")

plt.legend()    
plt.show()
```
